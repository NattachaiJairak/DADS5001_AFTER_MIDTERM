````python
import pandas as pd
````

````python
df_uploaded = pd.read_csv("https://raw.githubusercontent.com/NattachaiJairak/DADS5001_AFTER_MIDTERM/main/DADS5001_2024-03-16/DADS5001_2024-03-16_In%20Class_SB_directory.csv")
df_uploaded.head()
````

````python
import plotly.express as px

# ตรวจสอบรายชื่อคอลัมน์ทั้งหมดใน DataFrame
print(df_uploaded.columns)
````

````python
import plotly.express as px

# สร้าง scatter plot
fig = px.scatter(df_uploaded, x='Street Address', y='Store Name', title='Scatter Plot of Organizations')

# แสดง scatter plot
fig.show()
````

````python
import plotly.express as px

# สร้างแผนภูมิกล่อง
fig = px.box(df_uploaded, x='Brand', y='Latitude', title='Box Plot of Latitude by Brand')

# แสดงแผนภูมิกล่อง
fig.show()
````

````python
import plotly.express as px

# สร้างแผนภูมิฮิสโตแกรม
fig = px.histogram(df_uploaded, x='Longitude', title='Histogram of Longitude')

# แสดงแผนภูมิฮิสโตแกรม
fig.show()
````

````python
# Filling missing values in 'City' and 'Postcode' columns with placeholders
df_uploaded['City'].fillna('Unknown', inplace=True)
df_uploaded['Postcode'].fillna('N/A', inplace=True)

# Creating a treemap
fig = px.treemap(df_uploaded, path=['Country', 'State/Province', 'City'], values='Store Number')

# Displaying the treemap
fig.show()
````

````python
# Filling missing values in 'City' and 'Postcode' columns with placeholders
df_uploaded['City'].fillna('Unknown', inplace=True)
df_uploaded['Postcode'].fillna('N/A', inplace=True)

# Creating a sunburst chart
fig = px.sunburst(df_uploaded, path=['Country', 'State/Province', 'City'], values='Store Number')

# Displaying the sunburst chart
fig.show()
````

````python
import pandas as pd
import matplotlib.pyplot as plt

# List of countries you're interested in
countries = ['TH', 'MY', 'SG', 'ID']

# Filter the dataframe for the specified countries and count the stores
store_counts = df_uploaded[df_uploaded['Country'].isin(countries)]['Country'].value_counts()

# Creating the bar chart using Matplotlib
plt.figure(figsize=(10, 6))
store_counts.plot(kind='bar', color='skyblue')
plt.title('Number of Starbucks Stores in Thailand, Malaysia, Singapore, and Indonesia')
plt.xlabel('Country')
plt.ylabel('Number of Stores')
plt.xticks(rotation=45)
plt.show()
````

แนะนำให้ใช้ความคล้ายคลึงกับเรื่องที่คุณต้องการอธิบายในบรรทัดต่อไปต่อไปนี้:
1. **Import Libraries**: เริ่มต้นด้วยการนำเข้าไลบรารีที่จำเป็น เช่น pandas เพื่อการจัดการข้อมูลและ matplotlib.pyplot เพื่อสร้างกราฟ
```python
import pandas as pd
import matplotlib.pyplot as plt
```
2. **Specify Countries**: กำหนดรายการของประเทศที่สนใจไว้ในตัวแปร `countries`
```python
countries = ['TH', 'MY', 'SG', 'ID']
```
3. **Filter Data**: กรองข้อมูลเพื่อให้ได้เฉพาะข้อมูลของประเทศที่กำหนดไว้ใน `countries` และนับจำนวนร้านสตาร์บัคส์ในแต่ละประเทศ
```python
store_counts = df_uploaded[df_uploaded['Country'].isin(countries)]['Country'].value_counts()
```
4. **Create Plot**: สร้างแผนภูมิแท่งโดยใช้ข้อมูลจำนวนร้านสตาร์บัคส์ในแต่ละประเทศ โดยใช้ `plt.bar()` ของ matplotlib
```python
plt.figure(figsize=(10, 6))
store_counts.plot(kind='bar', color='skyblue')
```
5. **Customize Plot**: ปรับแต่งรายละเอียดต่างๆ ของแผนภูมิ เช่น ชื่อกราฟ และป้ายกำกับแกน
```python
plt.title('Number of Starbucks Stores in Thailand, Malaysia, Singapore, and Indonesia')
plt.xlabel('Country')
plt.ylabel('Number of Stores')
plt.xticks(rotation=45)
```
6. **Show Plot**: แสดงแผนภูมิที่สร้างขึ้น
```python
plt.show()
```
ด้วยขั้นตอนทั้งหมดนี้ คุณจะสามารถสร้างแผนภูมิแท่งที่แสดงจำนวนร้านสตาร์บัคส์ในแต่ละประเทศได้ โดยแสดงข้อมูลแบบอธิบายอย่างละเอียดในแต่ละขั้นตอนของโค้ด

````python
# Filter the dataframe for Starbucks stores in Thailand with specified postal codes (50000 to 50300)
thai_stores_postcodes = df_uploaded[
    (df_uploaded['Country'] == 'TH') &
    (df_uploaded['Postcode'].astype(str).str[:5].isin([str(i) for i in range(50000, 50301)]))
]

# Plotting the locations on a map using Plotly
fig = px.scatter_geo(thai_stores_postcodes,
                     lat='Latitude',
                     lon='Longitude',
                     hover_name='Store Name',
                     title='Locations of Starbucks Stores in Thailand with Postcodes 50000 to 50300',
                     scope='asia')

fig.show()
````

# กรองข้อมูลใน DataFrame เพื่อเลือกเฉพาะร้านสตาร์บัคส์ในประเทศไทยที่มีรหัสไปรษณีย์ที่ระบุ (50000 ถึง 50300)
thai_stores_postcodes = df_uploaded[
    (df_uploaded['Country'] == 'TH') &  # เลือกเฉพาะประเทศไทย
    (df_uploaded['Postcode'].astype(str).str[:5].isin([str(i) for i in range(50000, 50301)]))  # กรองโดยใช้รหัสไปรษณีย์
]

# สร้างแผนที่และแสดงตำแหน่งของร้านที่กรองแล้วบนแผนที่โดยใช้ Plotly Express
fig = px.scatter_geo(thai_stores_postcodes,  # ใช้ข้อมูลที่กรองแล้ว
                     lat='Latitude',  # ค่าละติจูด
                     lon='Longitude',  # ค่าลองจิจูด
                     hover_name='Store Name',  # ชื่อที่แสดงเมื่อโฮเวอร์
                     title='Locations of Starbucks Stores in Thailand with Postcodes 50000 to 50300',  # ชื่อของแผนที่
                     scope='asia')  # ขอบเขตของแผนที่

# แสดงแผนที่
fig.show()
