```python
import pandas as pd

# URL of the CSV file
url = "https://raw.githubusercontent.com/NattachaiJairak/DADS5001_AFTER_MIDTERM/main/DADS5001_2024-03-23/Data_Homework.csv"

# Read the CSV file into a DataFrame
df = pd.read_csv(url)

# Display the column names
print(df.columns)
```

```python
# 1. Remove leading and trailing whitespaces
df[' Budget Amount (THB) '] = df[' Budget Amount (THB) '].str.strip()

# 2. Remove commas from the values
df[' Budget Amount (THB) '] = df[' Budget Amount (THB) '].str.replace(',', '')

# 3. Convert the values to numeric data type
df[' Budget Amount (THB) '] = pd.to_numeric(df[' Budget Amount (THB) '], errors='coerce')

# Display the first few rows of the DataFrame after cleaning
df
```

```python
# Group by 'Cost center name', 'Project name', and 'Classificated', and sum 'Budget Amount (THB)'
grouped_df = df.groupby(['Cost center name', 'Project name', 'Classificated'])[' Budget Amount (THB) '].sum().reset_index()

# Display the resulting DataFrame
grouped_df
```

```python
import matplotlib.pyplot as plt

# Get unique values in 'Classificated' column
unique_classificated_values = df['Classificated'].unique()

# Plot separate line charts for each unique value in 'Classificated'
for value in unique_classificated_values:
    # Filter DataFrame for the current 'Classificated' value
    filtered_df = df[df['Classificated'] == value]

    # Create the line chart
    plt.figure(figsize=(10, 6))  # Adjust figure size if needed
    plt.plot(filtered_df['Cost center name'], filtered_df[' Budget Amount (THB) '], marker='o', linestyle='-')
    plt.title(f'Line Chart for {value}')
    plt.xlabel('Cost center name')
    plt.ylabel('Budget Amount (THB)')
    plt.xticks(rotation=90)  # Rotate x-axis labels if needed
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
import plotly.express as px

# Create Treemap
fig = px.treemap(grouped_df, path=['Classificated', 'Cost center name', 'Project name'], values=' Budget Amount (THB) ')
fig.show()
```
